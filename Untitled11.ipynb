```python
from math import exp
from math import erfc
import matplotlib
from matplotlib import pyplot as plt
import matplotlib.animation as animation
from matplotlib import cm
from jupyterthemes import jtplot
jtplot.style()
```

```python
def run():
    #f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
    ax1=plot_mat("water")
    ax2=plot_mat("air")
#1.562
def plot_mat(mat):   

    if mat=="water":
        Vis=8.90*10**-4    
        start=1600
        t_array=[1600]
    if mat=="air":
        Vis=0.1562*10**-4  
        start=102400
        t_array=[102400]
    
    for i in range(11):
        start=start*2
        t_array.append(start)
    #print(t_array)  
   # t_array=[100,5000,250000,12500000,625000000]
    norm = matplotlib.colors.Normalize(
        vmin=log(min(t_array),5)/log(max(t_array),5)-log(min(t_array),5)/log(max(t_array),5),
        vmax=log(max(t_array),5)/log(max(t_array),5)-log(min(t_array),5)/log(max(t_array),5))
    # choose a colormap
    c_m = matplotlib.cm.cool
    # create a ScalarMappable and initialize a data structure
    s_m = matplotlib.cm.ScalarMappable(cmap=c_m, norm=norm)
    s_m.set_array([])
    for t in t_array:   
        print((log(t,5)/log(max(t_array),5)))
        y=[i/50 for i in range(100*50)]
        U=1
        V_x=[U*(erfc(yval/(Vis*t))) for yval in y]
        plt.xlabel("y")
        plt.ylabel("V_x")
        plt.title("Plot of V_x against y for " +mat+ " while varying t")
        plt.plot(V_x,y,c=c_m(2*(log(t,5)/log(max(t_array),5)-log(min(t_array),5)/log(max(t_array),5))))
    plt.colorbar(s_m)
    #plt.colorbar()
    plt.show()
```

```python
run()
```
